# Batch Processing Method


```python
'''
1. Read a csv file into a DataFrame
2. Delete columns with NaN
3. Extract only the necessary columns to aggregate the loan counts and create a new DataFrame
4. Group the loan counts based on book title, author, ISBN, and volume using groupby.
5. Remove duplicate rows from the original DataFrame and extract only the unique rows to create a copy
6. Set ‘book title, author, ISBN, volume’ as the index
7. Update the original DataFrame with the cumulative loan counts stored in loan_count
8. Reset the index
9. Change to the column order of the original DataFrame
'''

def data_cleaning(filename):

  #1
  ns_df = pd.read_csv(filename, low_memory=False)

  #2
  ns_book = ns_df.dropna(axis=1, how='all')

  #3
  count_df = ns_book[['도서명','저자', 'ISBN', '권', '대출건수']]

  #4
  loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()

  #5
  dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy()

  #6
  ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)

  #7
  ns_book3.update(loan_count)

  #8
  ns_book4 = ns_book3.reset_index()

  #9.
  ns_book4 = ns_book4[ns_book.columns]

  return ns_book4

```

```python
new_ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.equals(new_ns_book4)
```
